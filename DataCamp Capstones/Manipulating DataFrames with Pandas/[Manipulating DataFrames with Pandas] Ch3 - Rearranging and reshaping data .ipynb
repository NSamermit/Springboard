```python
Video - Pivot tables
#Pivot limitations - Pivot requires unique index/column pairs to ID values in the new table

#Pivot tables reshapes a df much like .pivot() 
#by summarizing the df w/ a pair of summarizing variables and their values
#pivot tables deals with multiple values for the same index/column pair using a reduction
#by default the reduction is an average

#can enable other aggregation arguments using keyword aggfunc=
#aggfunc='count' is to count the number of values for each combination
    #this makes it a frequency table
```

```python
#Ex8 - Setting up a pivot table
'''pivot table allows you to see all of your variables as a function of two other variables. '''
#users df:
  weekday    city  visitors  signups
0     Sun  Austin       139        7
1     Sun  Dallas       237       12
2     Mon  Austin       326        3
3     Mon  Dallas       456        5

# Create the DataFrame with the appropriate pivot table: by_city_day
by_city_day = pd.pivot_table(users, 
                             index='weekday',
                             columns='city')

# Print by_city_day
print(by_city_day)
        signups        visitors       
city     Austin Dallas   Austin Dallas
weekday                               
Mon           3      5      326    456
Sun           7     12      139    237
```

```python
#Ex9 - Using other aggregations in pivot tables
#practice using len and count aggfuncs
# Use a pivot table to display the count of each column: count_by_weekday1
count_by_weekday1 = pd.pivot_table(users,
                                   index='weekday',
                                   aggfunc='count')

# Print count_by_weekday
print(count_by_weekday1)

# Replace 'aggfunc='count'' with 'aggfunc=len': count_by_weekday2
count_by_weekday2 = pd.pivot_table(users,
                                   index='weekday',
                                   aggfunc=len) #no ''

# Verify that the same result is obtained
print('==========================================')
print(count_by_weekday1.equals(count_by_weekday2))
#note how we generated boolean return using .equals()
         city  signups  visitors
weekday                         
Mon         2        2         2
Sun         2        2         2
==========================================
True
```

```python
#Ex10 - Using margins in pivot tables
#Useful to add totals in margins of a pivot table. Use margins=True
#learn new aggfunc: sum
#users df:
  weekday    city  visitors  signups
0     Sun  Austin       139        7
1     Sun  Dallas       237       12
2     Mon  Austin       326        3
3     Mon  Dallas       456        5

# Create the DataFrame with the appropriate pivot table: signups_and_visitors
signups_and_visitors = pd.pivot_table(users,
                                      index='weekday',
                                      aggfunc=sum)

# Print signups_and_visitors
print(signups_and_visitors)

         signups  visitors
weekday                   
Mon            8       782
Sun           19       376

# Add in the margins: signups_and_visitors_total 
signups_and_visitors_total = pd.pivot_table(users, 
                                            index='weekday', 
                                            aggfunc=sum, #without aggfunc the summation is wrong
                                            margins=True)

# Print signups_and_visitors_total
print(signups_and_visitors_total)
         signups  visitors
weekday                   
Mon            8       782
Sun           19       376
All           27      1158
```
