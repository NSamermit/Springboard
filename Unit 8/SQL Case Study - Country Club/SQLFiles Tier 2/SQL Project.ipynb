different costs to members (the listed costs are per half-hour 'slot'), and

facility, the name of the member formatted as a single column, and the cost.

SELECT
    f.name AS 'Reserved Facility',
    CASE
        WHEN m.memid = 0 THEN m.firstname
        ELSE CONCAT(m.firstname, ' ', m.surname)
        END as 'Guest Name',
    (CASE 
        WHEN b.memid = 0 THEN b.slots * f.guestcost
        ELSE b.slots * f.membercost
        END) AS cost,
    CONVERT(b.starttime, DATETIME) as date,
    EXTRACT(HOUR FROM CONVERT(b.starttime, DATETIME)) as 'Reservation Time'
FROM 
    Bookings as b
INNER JOIN 
    Members as m
ON 
    b.memid = m.memid
INNER JOIN 
    Facilities as f
ON 
    b.facid = f.facid
WHERE 
    EXTRACT(MONTH FROM CONVERT(starttime, DATE)) = 09
    AND
    EXTRACT(DAY FROM CONVERT(starttime, DATE)) = 14
    AND 
    CASE
        WHEN b.memid = 0 THEN b.slots * f.guestcost
        ELSE b.slots * f.membercost

ORDER BY

```python
import sqlite3
from sqlite3 import Error
import pandas as pd
from sqlalchemy import create_engine
```

```python
db_file = "sqlite_db_pythonsqlite.db"
conn = sqlite3.connect(db_file)

def get_connection():
    return conn

#got this nifty idea from this StackOVerflow page:
# https://stackoverflow.com/questions/61991050/connect-to-sqlite3-connection-using-sqlalchemy
engine = create_engine('sqlite://', creator = get_connection)

with engine.connect() as con:
    facilities_results = con.execute('SELECT * FROM Facilities')
    members_results = con.execute('SELECT * FROM Members')
    bookings_results = con.execute('SELECT * FROM Bookings')
    
    facilities_df = pd.DataFrame(facilities_results.fetchall())
    facilities_df.columns = facilities_results.keys()
    
    members_df = pd.DataFrame(members_results.fetchall())
    members_df.columns = members_results.keys()
    
    bookings_df = pd.DataFrame(bookings_results.fetchall())
    bookings_df.columns = bookings_results.keys()
    
```

```python
def query(arg):
    '''Simple function to input a query into the database
        The query returns as a dataframe
    '''
    with engine.connect() as con:
        rs = con.execute(arg)
        df = pd.DataFrame(rs.fetchall())
        df.columns = rs.keys()
        print(df.to_string(index=False))
```

```python
#Q10:Produce a list of facilities with a total revenue less than 1000.
#The output of facility name and total revenue, sorted by revenue. Remember
#that theres a different cost for guests and members!
```

```python
#Key Concept is that the SELECT x FROM Y is an iterative loop similar to python's
#for x in data_structure: where data_structure looks more like a dictionary
question10 = '''
SELECT 
    f.name as 'Facility', 
    (SELECT
    (SUM(CASE WHEN b.memid <> 0 AND (SELECT facid FROM Bookings) THEN 1 ELSE 0 END) -- can use to check for g:m reservation
        * (SELECT membercost FROM Facilities WHERE b.facid = facid)
        * (SELECT slots FROM Bookings)) -- This is the revenue from members
        +
        (SUM(CASE WHEN b.memid = 0 AND (SELECT facid FROM Bookings) THEN 1 ELSE 0 END) 
        * (SELECT guestcost FROM Facilities WHERE b.facid = facid) 
        * (SELECT slots FROM Bookings))) -- This is the revenue from guests
    AS Revenue   
FROM 
    Facilities as f
INNER JOIN 
    Bookings AS b 
USING (facid)
GROUP BY f.name
HAVING Revenue < 1000
ORDER BY Revenue
;
'''

query(question10)
```
